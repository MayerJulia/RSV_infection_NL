```r
# Load packages
library(mgcv)
library(ggplot2)
library(tidyverse)
library(lubridate)
library(gridExtra)
library(grid)
library(writexl)
library(magrittr)
library(rstatix)
```

```r
lambda_sp_nod <- read.csv("Lamda_sp_nod.csv")
lambda_sp_d <- read.csv("lamnda_spd.csv")
```

```r
ggplot() + geom_line(lambda_sp_nod, mapping = aes(x=agemid, y=median, color = 'Not attending daycare')) + 
           geom_line(lambda_sp_d, mapping = aes(x=agemid, y=median, color = 'Attending daycare')) + 
           geom_ribbon(lambda_sp_nod, mapping = aes(x=agemid, ymin=low95, ymax=up95), fill="blue", alpha=0.3) +
           geom_ribbon(lambda_sp_d, mapping = aes(x=agemid, ymin=low95, ymax=up95), fill="red", alpha=0.3) +
           xlab("Age (days)") + ylab("FOI") + labs(color = 'Daycare') + ggtitle("FOI for the spring birth cohort") 
```
